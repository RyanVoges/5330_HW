## _Part A: Scenario_
We used the Heating Oil data, and our scenario places us as the supplier of heating oil. That places us in the short hedge position. We want to short the futures contract to lock in a profitable selling price so we don't have to worry about market volitility. We assume we will deliver one gallon of heating oil per contract period, and that the period lasts for 20 working days. We also engage in a stack and roll strategy, which holds the hedge ratio constant for the following period.

## _Part B: Data Collection_
Note: Data will be submitted with the notebook. We chose to focus on the dates from January 2015 - December 2019.

Using the HOFuters.csv file and the HOSpot.csv file we combined the data using the Merge_Data_Code.ipynb file resulting in the Merged.csv file. We used the Merged.csv file for all of our analysis.

## _Part C: Coding_
Note: As you can see, our log estimate of beta is equivalent to our non-log estimate. Therefore, we see no distinction between the two methods for this particular data set, and our results are expressed in the non-log format.

```julia
using StatsKit
using StatsPlots
using DataFrames
```

```julia
df = DataFrame!(CSV.read("Merged.csv"));
```

```julia
first(df,5)
```

```julia
sort!(df.Date, rev=false);
```

```julia
M_Spot = convert(Array, df.HOSpot)
M_Futures = convert(Array, df.HOFutures);
```

```julia
StatsPlots.plot(M_Spot)
```

```julia
StatsPlots.plot!(M_Futures)
```

```julia
log_M_Spot = M_Spot
for i = 1:length(M_Spot)
    log_M_Spot[i] = log(M_Spot[i])
end
```

```julia
log_M_Futures = M_Futures
for i = 1:length(M_Futures)
    log_M_Futures[i] = log(M_Futures[i])
end
```

```julia
StatsPlots.plot(log_M_Spot)
```

```julia
StatsPlots.plot!(log_M_Futures)
```

```julia
N = length(M_Spot)
ΔS = diff(M_Spot)
ΔF = diff(M_Futures)
Δlog_S = diff(log_M_Spot)
Δlog_F = diff(log_M_Futures);
```

```julia
X = ones(N-1, 2);
log_X = ones(N-1, 2)
X[:,2] .= ΔF
y = reshape(ΔS, (N-1, 1))
log_X[:,2] .= Δlog_F
log_y = reshape(Δlog_S, (N-1, 1));
```

```julia
coefs = StatsKit.MultivariateStats.llsq(X[1:60,:], y[1:60]; bias = false)
log_coefs = StatsKit.MultivariateStats.llsq(log_X[1:60,:], log_y[1:60]; bias = false);
```

```julia
print(coefs[2])
print("\n")
print(log_coefs[2])
```

```julia
M = 60
hedgeratios = zeros(M)
ibeg = 1
iend = 60
ii = 1
stair = 20
for i = 1:stair:(stair*M)
    coefs = StatsKit.MultivariateStats.llsq(X[ibeg:iend,:], y[ibeg:iend]; bias = false) 
    hedgeratios[ii] = coefs[2]
    ibeg += stair
    iend += stair
    ii += 1
end
```

```julia
M = 60
log_hedgeratios = zeros(M)
ibeg = 1
iend = 60
ii = 1
stair = 20
for i = 1:stair:(stair*M)
    log_coefs = StatsKit.MultivariateStats.llsq(log_X[ibeg:iend,:], log_y[ibeg:iend]; bias = false) 
    log_hedgeratios[ii] = log_coefs[2]
    ibeg += stair
    iend += stair
    ii += 1
end
```

```julia
StatsPlots.plot(hedgeratios)
```

```julia
StatsPlots.plot!(log_hedgeratios)
```

```julia
avg = mean(hedgeratios)
size = length(hedgeratios)
hratios = map(x -> (x - avg)^2, hedgeratios)
added = sum(hratios)
std = sqrt(added / (size - 1))
iqrange = iqr(hedgeratios)
med = median(hedgeratios)
#quantiles
p_25 = quantile(hedgeratios, .25)
p_75 = quantile(hedgeratios, .75);
```

## _Part C continued: Summary Statistics_

```julia
print("*********Summary Stats*********")
print("\n")
print("\n")
print("Mean: ", avg)
print("\n")
print("Standard Deviation: ", std)
print("\n")
print("Inner Quartile Range: ", iqrange)
print("\n")
print("Median: ", med)
print("\n")
print("1st Quartile: ", p_25)
print("\n")
print("3rd Quartile: ", p_75)
```
